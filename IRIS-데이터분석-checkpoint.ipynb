```python
from sklearn.datasets import load_iris
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_graphviz
import graphviz
```

```python
iris_sns = sns.load_dataset('iris')
iris_sns.head()
```

```python
iris_sns.columns
```

```python
sns.pairplot(iris_sns, hue='species')
```

```python
x= iris.data[:, 2:]
y= iris.target
```

```python
y
```

```python
tree_clf = DecisionTreeClassifier(max_depth=2, random_state=12)
```

```python
tree_clf.fit(x,y)
```

```python
pred = tree_clf.predict([[1.4, 0.2], [5.0, 1.8]])
```

```python
iris.target_names[pred]
```

```python
iris_sns.tail(5)
```

```python
export_graphviz(
    tree_clf,
    out_file='data/iris.dot',
    feature_names=['petal length', 'petal width'], 
    class_names=iris.target_names,
    rounded=True,
    filled=True
)
```

```python
with open('data/iris.dot') as file:
    file_data = file.read()
file_data
```

```python
data = graphviz.Source(file_data)
data.format = 'png'
data.render(filename='iris_tree_clf', directory='images', cleanup=True)
```
