```python
bins = [0, 3, 8, 15, 30, 60, 100]
labels = ['baby', 'children', 'teenage', 'young', 'adult', 'old']
data['age_kind'] = pd.cut(data.age, bins=bins, labels=labels, include_lowest=True)
```

```python
data.head(10)
```

```python
data.groupby('age_kind').mean(numeric_only=True)
```

```python
plt.figure(figsize=(12, 6))
# 131 => nrow=1  ncol=3  index=1
plt.subplot(131)
sns.barplot(x='pclass', y='survived', data=data, ci=False)
plt.subplot(132)
sns.barplot(x='age_kind', y='survived', data=data)
plt.subplot(133)
sns.barplot(x='sex', y='survived', data=data)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(12,6))
sns.countplot(x='sex', data=data, ax=ax[0])
ax[0].set_title('성별 승객수')

sns.countplot(x='sex', hue='survived', data=data, ax=ax[1])
ax[1].set_title('성별 승객의 생존수')
```
