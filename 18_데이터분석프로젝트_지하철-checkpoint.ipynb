```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
subway_data = pd.read_csv('서울시 12월 지하철 데이터.csv')
```

```python
subway = subway_data.copy()
```

```python
subway.head()
```

```python
subway = subway.drop('aa', axis=1)
subway.head()
```

```python
subway = subway.drop('등록일자', axis=1)
subway.head()
```

```python
subway.shape
```

```python
subway.info()
```

```python
subway.describe()
```

```python
subway = subway.rename(columns = {'사용일자' : 'date',
                                  '노선명' : 'lane',
                                  '역명' : 'station',
                                  '승차총승객수' : 'getin',
                                  '하차총승객수' : 'getout'
                                 })
```

```python
subway.head()
```

```python
subway['lane'].dtypes
```

```python
subway['lane'].value_counts()
```

```python
lane_sum = subway.dropna(subset=['getin']).groupby('lane', as_index=False).agg(lane_getin=('getin','sum'))
```

```python
lane_sum.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 10))
plt.rcParams.update({'font.family':'Malgun Gothic'})
plt.xticks(rotation=60)

sns.barplot(data=lane_sum, x='lane', y='lane_getin')
plt.show()
```

```python
subway[['station', 'getin', 'getout']].head()
```

```python
subway_lane8 = subway[subway['lane']=='8호선']
```

```python
subway_lane8.head(17)
```

```python
subway_lane8['station'] = subway_lane8['station'].replace({'잠실(송파구청)' : '잠실',
                                                           '천호(풍납토성)' : '천호',
                                                           '몽촌토성(평화의문)' : '몽촌토성',
                                                           '남한산성입구(성남법원.검찰청)' : '남한산성'
                                                          })
```

```python
station_getin = subway_lane8.dropna(subset=['getin']).groupby('station',
                                    as_index=False).agg(getin_sum=('getin','sum'))
```

```python
station_getin
```

```python
plt.figure(figsize=(12, 6))


sns.barplot(data=station_getin, x='station', y='getin_sum')
plt.title('8호선 역의 12월 총 승객 수')
plt.xlabel('역명')
plt.xticks(rotation=60)
plt.ylabel('승차 총 승객 수')

plt.show()
```
